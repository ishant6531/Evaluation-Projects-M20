# Exploratory data analysis

```python
# Ploting barplot of students marks in Math subject.
plt.figure(figsize=(10,6))
pd =sns.barplot(x = "Name", y = "Maths",  data=df)
for p in pd.patches:
    pd.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=10,
                   xytext = (0, 8), 
                   textcoords = 'offset points')

plt.xlabel(" Students Names " , size=15)
plt.ylabel("Marks " ,size=15)
plt.title (" Students  Marks in Math " , size=15)
plt.show()
# conclusion :-  Mukesh got highest marks and ramesh got lowest marks.
```

```python
# Making countplot  of total no of males and females.
plt.figure(figsize=(8,5))
pd =sns.countplot(x = "Gender" ,data=df )
for p in pd.patches:
    pd.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=18,
                   xytext = (0, -20), 
                   textcoords = 'offset points')
plt.xlabel("Gender", size=15)
plt.ylabel("count" ,size=15)
plt.title ("  Attrition  Rate with Gender  " , size=15)
plt.show()
# conclusion - No of males and females is same.
```
