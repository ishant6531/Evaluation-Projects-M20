# Importing Necessary libraries

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
```

```python
def maxr2_score(regr,x,y):
    max_r_score=0
    for i in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state= i ,test_size=0.20)
        regr.fit(x,y)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_i=i
    print(" We are getting maximum  r2 score corresponding to",final_i,"is",max_r_score)
    return final_i
```

```python
#Finding best random state value for Linear Regression
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
r_state=maxr2_score(lreg,x,y)
```

```python
from sklearn.model_selection import cross_val_score
print("Mean r2 score for Linear Regression:",cross_val_score(lreg,x,y,cv=5,scoring="r2").mean())
print("standard deviation in r2 score for Linear Regression",cross_val_score(lreg,x,y,cv=5,scoring="r2").std())
print()
```

# Using GridSearchCv to find best parameters of different algorithims

```python
sv=SVR()
parameters={'kernel':('linear','rbf','poly','sigmoid'),'C':[1,10]}
gd=GridSearchCV(sv,parameters)
gd.fit(x,y)
print(gd.best_params_)
print("\n")


knn=KNeighborsRegressor()
parameters={'n_neighbors':(5,20)}
gd=GridSearchCV(knn,parameters)
gd.fit(x,y)
print(gd.best_params_)
print("\n")

dtc=DecisionTreeRegressor()
parameters={'criterion':('mse', 'friedman_mse', 'mae')}
gd=GridSearchCV(dtc,parameters)
gd.fit(x,y)
print(gd.best_params_)
print("\n")


lsreg=Lasso()
parameters={"alpha":[0.001,0.01,0.1,1]}
gd=GridSearchCV(lsreg,parameters)
gd.fit(x,y)
print(gd.best_params_)
print("\n")

from sklearn.linear_model import ElasticNet
enr=ElasticNet()
parametors={"alpha":[0.001,0.01,0.1,1]}
gd=GridSearchCV(enr,parametors,cv=10)
gd.fit(x,y)
gd.best_params_
```
